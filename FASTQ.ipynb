# IMPORTS

```python
import numpy as np
from termcolor import colored
import random
from itertools import permutations
```

# Storing Protein and PHRED scores in one variable

```python
S=[] #storing the individual DNA strings
P=[] #storing corresponding quality score for each DNA character
for i in seq[1::4]:
  S.append(i)
for j in seq[5::4]:#FINAL STRING

  P.append(j)
```

```python
for i in range(len(S)):
  S[i]=S[i].strip('\n')
final_S=''
for i in S:
  final_S+=i
print(len(final_S))
```

```python
#storing corresponding base cell accuracy for each DNA character
l_bca=[]

for i in range(len(P)):
  l=[]
  for v in P[i]:
    if v not in "\n":
      v=ord(v)-33
      l.append(round(100-(10**((-1)*int(v)/10)*100),2))
  l_bca.append(l)
t=[]
for i in l_bca:
  for j in i:
    t.append(j)
print(t)
```

# NumPy Array of PHRED base cell accuracies

```python
#list as a NumPy array

final_P=np.array(t)

print(final_P)
```

# Function to generate RANDOM ministring

```python
"""
@param n: user defined length of ministring 
"""

def generate_mini(n):
  ministr=''
  for j in range(n):
    ministr+=random.choice("ATGC")
  return (ministr)


length_mini=int(input('Enter preferred length of Mini string : ')) #length of ministring
ministr=generate_mini(length_mini) #ministring 
print(ministr)
```

# Task 1

```python
"""
Matching the dna sequences directly in order to distinguish between them based on base cell accuracy. 

@param r: Master string
@param m: Randomly generated mini string of user defined length
@param p: Array of PHRED quality scores
@param t: Threshold value for accuracy of match

"""

def directMatch(r, m, p,t):
  temp=[]
  results=[]
 
  d=r.count(m)
  print('Number of matches : ', d)

  i=-1
  while d!=0:
    index=r.find(m,i+1)
    i=index
    if index not in temp:
      temp.append(index)
    d=d-1

  check=True
  for j in temp:
    bsa=str(p[j:j+length_mini])
    for k in p[j:j+length_mini]:
      if k>=t:
        check=True;
      else:
        check=False
        break
    if check==False:
        results.append(colored('\nFor the substring match from loc '+str(j)+' having BCA as : '+bsa+' the match is INNACURATE','red',attrs=['bold', 'underline']))
    else:
        results.append(colored('\nFor the substring match from loc '+str(j)+' having BCA as : '+bsa+' the match is ACCURATE','green',attrs=['bold', 'underline']))

  return results
```

```python
mainstr=final_S #Master string containing all DNA sequences
threshold=float(input('Enter threshold for accurate protein match: ')) #threshold for accurate protein match
direct = directMatch(mainstr, ministr, final_P,threshold) # calling the function directMatch

#print(colored("The Main String is : \n"+mainstr+'\n','yellow'))
print(colored("The Mini String is : \n"+ministr+'\n','blue'))
print(colored("The Base cell acc String is : \n"+str(final_P)+"\n",'magenta'))
print(colored('\n'+"*"*50+ '\n\nResults: ',attrs=['bold','underline']))
for i in direct:
    print(colored(i,attrs=['bold','underline']))
```

# Task 2

```python
"""
Matching the dna sequences directly in order to distinguish between them based on MEAN base cell accuracy.
@param r: Master string
@param m: Randomly generated mini string of user defined length
@param p: Array of PHRED quality scores
@param t: Threshold value for accuracy of match

"""

def mean_directMatch(r, m, p,t):
  #global final_P
  temp=[]
  d=r.count(m)
  results2=[]
  length_mini=len(ministr) #length of ministring

  i=-1
  while d!=0:
    index=r.find(m,i+1)
    i=index
    if index not in temp:
      temp.append(index)
    d=d-1
    
  check=True
  for j in temp:
        sum=0
        for k in p[j:j+length_mini]:
            sum+=k
        avg=round(sum/length_mini,2)
        if avg>=t:
            results2.append(colored('\nFor the substring match from loc '+str(j)+' having mean BCA as : '+str(avg)+' the match is ACCURATE','green',attrs=['bold', 'underline']))
        else:
            results2.append(colored('\nFor the substring match from loc '+str(j)+' having mean BCA as : '+str(avg)+' the match is INNACURATE','red',attrs=['bold', 'underline']))
         
  return results2
```

```python
#threshold shd change accordingly : as it MEAN base cell accuracy
threshold_2=float(input('Enter threshold for MEAN accurate protein match: ')) #threshold for accurate protein match
direct = mean_directMatch(mainstr, ministr, final_P,threshold_2) # calling the function directMatch

#print(colored("The Main String is : \n"+mainstr+'\n','yellow'))
print(colored("The Mini String is : \n"+ministr+'\n','blue'))
print(colored("The Base cell acc String is : \n"+str(final_P)+"\n",'magenta'))
print(colored('\n'+"*"*50+ '\n\nResults: ',attrs=['bold','underline']))
for i in direct:
    print(colored(i,attrs=['bold','underline']))
```
